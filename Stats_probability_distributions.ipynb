```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set()
sns.set_context('notebook') 
```

# Normal Distribution
### Generate Random values

```python
# if you don't want to get the same random number again then,
np.random.seed(0)
x_rand = np.random.normal(loc=0, scale=1.0, size=1000) # loc=mean, standard deviation=scale
```

### Visualization

```python
# using matplotlib
plt.hist(x_rand, bins=20)
plt.show()
```

```python
# using seaborn
sns.histplot(x_rand, kde=False, bins =20)
# sns.histplot(data=x_rand, kde=False)
```

```python
sns.histplot(x_rand, kde=False, bins =100)
```

```python
sns.distplot(x_rand)
plt.show()
```

```python
sns.distplot(x_rand, kde=False)
plt.show()
```

```python
sns.distplot(x_rand, fit=stats.norm)
```

```python
sns.distplot(x_rand, fit=stats.norm, kde=False)
```

### Distribution Fitting

```python
stats.norm.fit(x_rand)
```

```python
loc, scale = stats.norm.fit(x_rand)
loc, scale
```

### Probability Density Function(PDF)

```python
# linearly spaced values of x
x = np.linspace(-5,5,100)
# np.linspace(start=-5, stop=5, num=100)
```

```python
# pdf
pdf = stats.norm.pdf(x, loc=loc, scale=scale)
```

```python
# plot
plt.plot(x,pdf, color='red')
plt.xlabel('x')
plt.ylabel('PDF')
plt.show()
```

### Cumulative Distribution Function

```python
# cdf
cdf = stats.norm.cdf(x, loc=loc, scale=scale)
```

```python
# plot
plt.plot(x, cdf, color='black')
plt.xlabel('x')
plt.ylabel('cdf = P(x<=X)')
```

### Percent Poit Function (PPF)

```python
# linearly spaced values of cdf
cdf_ = np.linspace(start=0, stop=1, num=10000)
```

```python
# X
x_ = stats.norm.ppf(cdf_, loc=loc, scale=scale)
```

```python
# plot
plt.plot(x_, cdf_, color='blue')
plt.xlabel('x')
plt.ylabel('cdf = P(x<=X)')
```

# Use of CDF

```python
# Probability that the variable X takes values lower than 1
x = np.linspace(-5,5,100)
cdf_1 = stats.norm.cdf(1, loc=loc, scale=scale)
cdf_1
```

```python
# plot
plt.plot(x, cdf, color='red')
plt.vlines(1, 0, cdf_1, linestyles=':') # x, ymin, ymax, colors=None, linestyles='solid'
plt.hlines(cdf_1, -5, 1, linestyles=':') # y, xmin, xmax, colors=None, linestyles='solid'
plt.xlabel('x')
plt.ylabel('cod = P(x<=X)')
```

```python
# value x that is not exceded 99% of the time
x_99 = stats.norm.ppf(0.99, loc=loc, scale=scale)
x_99
```

```python
# plot
plt.plot(x, cdf, color='red')
plt.vlines(x_99, 0, 0.99, linestyles=':')
plt.hlines(0.99, -5, x_99, linestyles=':')
plt.xlabel('x')
plt.ylabel('cod = P(x<=X)')
```
